```python
import chromadb
```

```python
client = chromadb.Client()
```

```python
client.create_collection(name="news")
```

```python
collection=client.get_collection("news")
```

```python
collection.add(
    documents=[
        "This is a document about pineapple",
        "This is a document about oranges",
        "This document is about apples",
        "This document is abput carrots"
    ],
    ids=["id1","id2","id3","id4"]
)
```

```python
collection.query(
    query_texts=["give me most citrus fruit"],
    n_results=2
)
```

```python
collection.update(
    documents=["This is about lemon"]
,
 ids=["id3"]
)
```

```python
collection.query(
    query_texts=["give me most citrus fruit"],
    n_results=2
)
```

```python
collection.delete(["id2"])
```

```python
collection.peek()
```
