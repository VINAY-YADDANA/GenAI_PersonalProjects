```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_groq import ChatGroq
llm=ChatGroq(model="llama-3.3-70b-versatile",temperature=1.8)
```

```python
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import JsonOutputParser
```

```python
prompt=''' 

From the below news article extrach revenue and eps in JSON format containint the
following keys: 'revenue_actual', 'revenue_expected', 'eps_actual','eps_expected'.capitalize
Each value should have a value million or billion
Only return the valid JSON. No Preamble

Article
========
{article}

'''
pt=PromptTemplate.from_template(prompt)
```

```python
chain = pt|llm
```

```python
article_text='''
Tesla
 reported third-quarter earnings Wednesday that topped analysts’ estimates even as revenue came in just shy of expectations. The stock popped roughly 17% in Thursday morning trading.

Here’s what the company reported compared with what Wall Street was expecting, based on a survey of analysts by LSEG:
Earnings per share: 72 cents, adjusted vs. 58 cents expected
Revenue: $25.18 billion vs. $25.37 billion expected
'''

response=chain.invoke({'article': article_text})
response
```

```python
response.content
```

```python
parser=JsonOutputParser()
output_parser=parser.parse(response.content)
output_parser
```
