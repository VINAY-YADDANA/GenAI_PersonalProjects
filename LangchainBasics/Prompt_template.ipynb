```python
from langchain_core.prompts import PromptTemplate

pt=PromptTemplate.from_template("Write two line poem on {topic1} and {topic2}")
#pt.invoke({"topic":"cat"})
```

```python
from langchain_groq import ChatGroq

llm=ChatGroq(model="openai/gpt-oss-120b",temperature=1.8)
llm.invoke("Hi")
```

```python
chain= pt|llm
response = chain.invoke({'topic1':'Cat','topic2':'bottle'})
result=response.content.replace("\n","").capitalize()
result.strip("-")
```

```python

from langchain_core.prompts import PromptTemplate

pt2=PromptTemplate.from_template("Write two line poem on {topic1}")
#pt.invoke({"topic":"cat"})


topics = ["Cat", "Bottle", "Tree"]

chain2=pt2|llm

results = []
for t in topics:
    response = chain2.invoke({'topic1': t})
    processed = response.content.replace("\n", "").capitalize().strip("-")
    results.append(processed)

print(results)
```
